##Line Plots


```python
y = arange(-5, 5, 0.01) #creates an array of numbers from -5 to 5 in steps of 0.01
E = cosh(y) #creates an array the same length as y of the cosh of each value in y
p = sinh(y) #creats an array of the sinh of each value of y
```

```python
plot(y, E, 'r-', linewidth = 2, label = "$equation 1$") #plots the first equation
plot(y, p, 'b:', linewidth = 2.5, label = "$equation 2$") #plots the second equation
ax = gca() #gets the current axes information
ax.spines['right'].set_color('none') #sets the color of the right axe to colorless
ax.spines['left'].set_color('none') #sets the left to colorless
ax.spines['top'].set_color('none') #sets the top axe to colorless
ax.spines['bottom'].set_color('none') #sets the bottom axe to colorless
axhline(y=0, color = "black") #sets the yaxis to the center and changes its color to black
axvline(x=0, color = "black") #sets the xaxis to the center and changes its color to black
title("Title of Graph", fontsize = 20) #turns on the title to size 20
xlabel("X label", fontsize = 14) #turns on x label to size 14
ylabel("Y label", fontsize = 14) #turns on y label to size 14
grid(True) #turns on the grid
legend() #turns on the legend
show() #tells the program to show the figure (prints it out)
```

##Subplots

```python
subplot(211) #this assigns the following plot to be the first subplot in the two row, one column array of subplots
plot(y, E, 'r-', linewidth = 2, label = "$equation 1$") #plots the first equation
ax = gca() #gets the current axes information
ax.spines['right'].set_color('none') #sets the color of the right axe to colorless
ax.spines['left'].set_color('none') #sets the left to colorless
ax.spines['top'].set_color('none') #sets the top axe to colorless

axhline(y=0, color = "black") #sets the yaxis to the center and changes its color to black
axvline(x=0, color = "black") #sets the xaxis to the center and changes its color to black
title("Graph 1", fontsize = 20) #turns on the title to size 20
xlabel("X 1", fontsize = 14) #turns on x label to size 14
ylabel("Y 1", fontsize = 14) #turns on y label to size 14
grid(True) #turns on the grid
legend(loc = "upper center") #turns on the legend
show() #tells the program to show the figure (prints it out)


subplot(212) #this assigns the following plot to the second subplot
plot(y, p, 'b:', linewidth = 2.5, label = "$equation 2$") #plots the second equation
ax = gca() #gets the current axes information
ax.spines['right'].set_color('none') #sets the color of the right axe to colorless
ax.spines['left'].set_color('none') #sets the left to colorless
ax.spines['top'].set_color('none') #sets the top axe to colorless
ax.spines['bottom'].set_color('none') #sets the bottom axe to colorless
axhline(y=0, color = "black") #sets the yaxis to the center and changes its color to black
axvline(x=0, color = "black") #sets the xaxis to the center and changes its color to black
title("Graph 2", fontsize = 20) #turns on the title to size 20
xlabel("X 2", fontsize = 14) #turns on x label to size 14
ylabel("Y 2", fontsize = 14) #turns on y label to size 14
grid(True) #turns on the grid
legend(loc = "lower right") #turns on the legend
show() #tells the program to show the figure (prints it out)
```

##Scatter Plots

```python
n = 200
x = random.normal(0,1,n)
y = random.normal(0,1,n) #Creates random data for you to plot

scatter(x, y, s=5, color = "red", label = "$Your Data$")
title("Graph Title", size = 16)
ylabel("Y label", size = 14)
xlabel("X label", size = 14)

xlim(xmax = 2, xmin = -2)
ylim(ymax = 2, ymin = -2)

grid(True)
legend()
show()
```

##Scatter Subplots

```python
n = 200
x = random.normal(0,1,n)
y = random.normal(0,1,n) #Creates random data for your first plot

a = random.normal(0,1,n)
b = random.normal(0,1,n) #Creates random data for your second plot

subplot(121)
scatter(x, y, s=5, color = "red", label = "$Your Data$")
title("Graph 1", size = 16)
ylabel("Y 1", size = 14)
xlabel("X 1", size = 14)
grid(True)
legend()

subplot(122)
scatter(a, b, s=5, color = "blue", label = "$Your Data$")
title("Graph 2", size = 16)
ylabel("Y 2", size = 14)
xlabel("X 2", size = 14)
grid(True)
legend()
```
