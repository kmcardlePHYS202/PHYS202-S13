```python
def is_anagram(s1,s2):
    if len(s1) != len(s2):
        return False
    one = []
    two = []
    for i in s1:
        one.append(i)
    for i in s2:
        two.append(i)
    anagram = False
    for i in one:
        for k in two:
            if i == k:
                anagram = True
                break
            anagram = False
    return anagram

print is_anagram("hello","olelh")
print is_anagram("hello","oleh")
print is_anagram("hello","hhhhh")
print is_anagram("this","siht")            
        
```
