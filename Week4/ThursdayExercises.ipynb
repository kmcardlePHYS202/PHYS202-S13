```python
x = arange(-0.01,0.01,0.001)
y = arange(-0.01,0.01,0.001)
X,Y = meshgrid(x,y)

q = -2e-9 #-2 nC
posx = 0
posy = -0.5e-3
V = pointPotential(X,Y,q,posx,posy)

fig = figure()
ax = fig.gca(projection='3d')

p = ax.plot_surface(X,Y,V, rstride=1, cstride=1, linewidth=0, cmap=cm.Reds)
show()
```
