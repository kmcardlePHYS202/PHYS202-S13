##Plotting the electric dipole potential in 3D
\begin{equation}
V(x,y)=\frac{kq}{\sqrt{x^2+(y-d)^2}}-\frac{kq}{\sqrt{x^2+(y+d)^2}}
\end{equation}

k=$1/4\pi\epsilon$

make sure to define variables when using natural units (when we say k=1, code that k=1)

```python
#make the x-y grid to cover the plane -2<x<2 and -2<y<2
dx = 0.06
dy = 0.06
x = arange(-2.,2.,dx)
y = arange(-2.,2.,dy)
X,Y = meshgrid(x,y)

#make the function
k = 1; q = 1;d = 1
V = (k*q/(sqrt(X**2+(Y-d)**2)))-(k*q/(sqrt(X**2+(Y+d)**2)))

#time to plot!
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')

p = ax.plot_surface(X,Y,V, rstride=1, cstride=1, linewidth=0, cmap=cm.seismic)
#help(Axes3D.plot_surface)
cb = fig.colorbar(p, shrink=0.5)
```

```python
#represent 3D data by contours
levels = array([16,8,4,2,1,0.5,0.25,0,-0.25,-0.5,-1,-2,-1,-4,-8,-16])
contour(X,Y,V,levels, cmap=cm.Set2_r, linewidths=1)
show()
```

```python
dx = 8./101
dy = 8./101
x = arange(-4.,4.,dx)
y = arange(-4.,4.,dy)
X,Y = meshgrid(x,y)

k = 1.3806488e-23 #Nm^2/C^2
q1 = -1e-9 #C
d = 2 #meters
q2 = 1e-9 #C
V = (k*q1/(sqrt((X-d)**2+(Y)**2)))+(k*q1/(sqrt((X+d)**2+(Y)**2)))+(k*q2/(sqrt((X)**2+(Y-d)**2)))+(k*q2/(sqrt((X)**2+(Y+d)**2)))

fig1 = figure()
ax1 = fig1.gca(projection='3d')

p = ax1.plot_surface(X,Y,V, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
cb = fig1.colorbar(p, shrink=0.5)

```
