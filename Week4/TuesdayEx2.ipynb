##Exercise1
Make a 3D plot of the function $G(x,y)=e^{-x^2-y^2}$  over range -6$\lt$$(x,y)$$\lt$6 using 51 bins in $x$ and 51 bins in $y$

```python
a = arange(-6,6,12./51.)
b = arange(-6,6,12./51.)

A,B = meshgrid(a,b)

G = exp(-A**2-B**2)

print G
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')

p = ax.plot_surface(A,B,G, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
cb = fig.colorbar(p, shrink=0.5)
cset = ax.contour(A,B,G, arange(0, 0.8, 0.2), zdir='x')
```

```python
dx = 8./101
dy = 8./101
x = arange(-4.,4.,dx)
y = arange(-4.,4.,dy)
X,Y = meshgrid(x,y)

k = 1.3806488e-23 #Nm^2/C^2
q1 = -1e-9 #C
d = 2 #meters
q2 = 1e-9 #C
V = (k*q1/(sqrt((X-d)**2+(Y)**2)))+(k*q1/(sqrt((X+d)**2+(Y)**2)))+(k*q2/(sqrt((X)**2+(Y-d)**2)))+(k*q2/(sqrt((X)**2+(Y+d)**2)))

fig1 = figure()
ax1 = fig1.gca(projection='3d')

p = ax1.plot_surface(X,Y,V, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
cb = fig1.colorbar(p, shrink=0.5)

```

```python
levels = array([-2e-31, -1e-31, -0.5e-31, -0.25e-31, -0.125e-31, 0, 0.125e-31, 0.25e-31, 0.5e-31, 1e-31, 2e-31])
contour(X,Y,V, levels, cmap=cm.Set2_r, linewidths=2)
show()
```
