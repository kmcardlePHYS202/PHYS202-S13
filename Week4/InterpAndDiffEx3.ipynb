```python
def finiteDifference(x,y):
    """finds the derivative of the function y over the range x"""
    dydx = zeros(y.shape,float)
    dydx[:-1] = diff(y)/diff(x)
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx

x = arange(0,10.1,0.1)
y = x**5

dydx1 = finiteDifference(x,y)

#actual derivative
actual = 5*x**4

plot(x,dydx1, "r")
plot(x,actual, "b")
show()
```

```python
x2 = arange(0,10.001, 0.001)
y2 = y = x2**5
dydx2 = finiteDifference(x2,y)
actual2 = 5*x2**4
percent_error = abs(actual2-dydx2)/actual2
plot(x2,percent_error)
ylim(ymin=0, ymax=0.01)
title("Percent Error")
show()
print max(percent_error)
```

its even more accurate when the step size is smaller!!!

```python
def fourPtFiniteDiff(x,y):
    """finds the derivative of the function y over the range x"""
    dydx = zeros(y.shape,float)
    for i in range(2,len(y)-2):
        a = y[i-2]
        b = y[i-1]
        c = y[i+1]
        d = y[i+2]
        dydx[i] = (a-8*b+8*c-d)/(12*(x[1]-x[0]))
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    dydx[-2] = (y[-2]-y[-3])/(x[-2]-x[-3])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[1] = (y[2]-y[1])/(x[2]-x[1])
    return dydx

x = arange(0,10.1,0.1)
y = x**5
dydx3 = fourPtFiniteDiff(x,y)
plot(x,dydx3, "r", linewidth=3)
plot(x,actual, "b--", linewidth=1.5)
show()

percent_error = abs(dydx3-actual)/actual
plot(x,percent_error)
ylim(ymin=0, ymax=0.01)
show()
```
