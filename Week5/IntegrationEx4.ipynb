```python
f = lambda t: exp(-t**2)
```

```python
N = 1000
a = 0
x = arange(0,3.1,0.1)
i = 0
E = zeros(len(x))
for b in x:
    h = (b-a)/N
    s = 0.5*f(a) + 0.5*f(b)
    for k in range(1,N):
        s += f(a+k*h)
    E[i]= (h*s)
    i+=1
    
```

```python
plot(x,E, 'r', linewidth = 2.5, label = "$E(x) = \int^x_0 e^{-t^2}dt$")
xlabel("$x$", size = 14)
ylabel("$E(x)$", size = 14)
title("Numeric Integration of $E(x) = \int^x_0 e^{-t^2}dt$\n\n", size = 15)
legend(loc = "lower right")
show()
```

```python
import scipy.integrate as integrate

n = 1 #ground state
A = lambda L: sqrt(2/L)
k = lambda L: (n*pi)/L
psi = lambda x,L: A(L)*sin(k(L)*x)
    
a = lambda L: L/3.
b = lambda L: L/2.

N = 10000000
x = arange(a(1),b(1)+0.01, (b(1)-a(1))/N)

Q = lambda y: (psi(y,1))**2

P = integrate.trapz(Q(x),x)
print "trapz:",P

P = integrate.quad(Q,1./3.,1./2.)
print "quad:",P[0]

x_all = arange(0,1.01,0.01)
plot(x_all,(psi(x_all,1))**2)
show()
```
