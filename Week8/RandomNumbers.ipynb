#Example 1

```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.
    
plot(x,f(x),label='True distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

```python
#number of samples
N=10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda z:arccos(cos(0)-2*z)

#sampled x values
xsamples = Finv(z);

#how do they compare to the original
hist(xsamples,50,normed=True,label='Samples distribution')
plot(x,f(x),label='True distribution',lw=3,color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

#Exercise 1

$F(x) = \int_{0}^x \lambda e^{-\lambda t}dt = -e^{-\lambda x} + e^{0}$

$F^{-1}(x)=\frac{ln(1-x)}{\lambda}$

```python
f = lambda x: lam*exp(-lam*x)
F = lambda x: -exp(-lam*x)+1
Finv = lambda x: log(1-x)/(-lam)

lam = 1
x = arange(0,5.01,0.01)
plot(x,f(x), label = 'True Distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)
legend(loc = 0)
show()
```

```python
#number of samples
N=10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#sampled x values
xsamples = Finv(z);

#how do they compare to the original
hist(xsamples,50,normed=True,label='Samples distribution')
plot(x,f(x),label='True distribution',lw=3,color='red')
show()
```
