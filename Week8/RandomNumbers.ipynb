#Example 1

```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.
    
plot(x,f(x),label='True distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

```python
#number of samples
N=10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda z:arccos(cos(0)-2*z)

#sampled x values
xsamples = Finv(z);

#how do they compare to the original
hist(xsamples,50,normed=True,label='Samples distribution')
plot(x,f(x),label='True distribution',lw=3,color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

$F(x) = \int_{0}^x \lambda e^{-\lambda t}dt = -e^{-\lambda x} + e^{0}$

$F^{-1}(x)=\frac{ln(1-x)}{-\lambda}$

```python
def distribute1D(x,prob,N):
    """takes any numerical distribution prob, on the intercal defined by the array x, and returns an array of N sampled values that are statistically the same as the input data"""
    dist = prob(x)
    CDF = cumsum(dist)/sum(dist)
    z = numpy.random.random(N)
    Finv = x[searchsorted(CDF,z)]
    return Finv

test = distribute1D(x,f,N)
hist(test,50,normed=True,label='Samples distribution')
show()
```

#Exercise 3

```python
d = 15*(10**-6) #15 micro meters
a = 10*(10**-6) #10 micro meters
L = 1 #1 meter
lam = 632.8*(10**-9) #632.8 nano meters

sinc = lambda x: sin(pi*x)/(pi*x)
I_diff = lambda x: sinc((a*x)/(lam*L))**2
I_int = lambda x: cos((2*pi*d*x)/(lam*L))**2
I_tot = lambda x: I_diff(x) * I_int(x)

N = 10000 #10,000 photons! 
x = arange(-0.2,0.201,0.001)
this = distribute1D(x,I_tot,N)
n,bins,patches=hist(this,400,normed=True,label='Samples distribution')
plot(x,I_tot(x)*max(n),'r')
xlim(-0.2,0.2)
show()
```
