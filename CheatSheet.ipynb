```python
x = arange(-pi,pi,0.01)
u = -cosh(x)**-2
i = 0+1j
l = [0, 0.5, 1.0, 2.0]
line = ['r-', 'r--', 'r-.', 'r:']
m = 0
for k in l:
    psik = ((i*k-tanh(x))/(i*k+1))*exp(i*k*x)
    plot(x, psik*psik.conjugate(), line[m], linewidth = 2.5, label = "$\psi^2, k=%s$" %k)
    m+=1
plot(x, u, linewidth = 2.5, label = "$U(x)$")

xlabel("$x(rad)$", fontsize = 15)
ylabel("$U(x), \psi^2(x)$", fontsize = 15)

grid(True)
xticks(arange(-pi, pi+1, pi/2), ["$-{\pi}$", "$-{\pi}/2$", "$0$", "$+{\pi}/2$", "$+{\pi}$"])
legend(loc = 'lower left',prop={'size':8})
show()
```

```python
from scipy.interpolate import interp1d

def finiteDifference(x,y):
    """finds the derivative of the function y over the range x"""
    dydx = zeros(y.shape,float)
    dydx[:-1] = diff(y)/diff(x)
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx

x = arange(0,10.1,0.1)
y = x**5

dydx1 = finiteDifference(x,y)

#actual derivative
actual = 5*x**4

plot(x,dydx1, "r", linewidth=3, label="computed derivative")
plot(x,actual, "b--", linewidth=3, label="actual derivative")
legend(loc = "upper left")
show()
```

```python
def fourPtFiniteDiff(x,y):
    """finds the derivative of the function y over the range x"""
    dydx = zeros(y.shape,float)
    for i in range(2,len(y)-2):
        a = y[i-2]
        b = y[i-1]
        c = y[i+1]
        d = y[i+2]
        dydx[i] = (a-8*b+8*c-d)/(12*(x[1]-x[0]))
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    dydx[-2] = (y[-2]-y[-3])/(x[-2]-x[-3])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[1] = (y[2]-y[1])/(x[2]-x[1])
    return dydx

x = arange(0,10.1,0.1)
y = x**5
dydx3 = fourPtFiniteDiff(x,y)
plot(x,dydx3, "r", linewidth=3, label="computed derivative")
plot(x,actual, "b--", linewidth=3,label="actual derivative")
legend(loc = "upper left")
title("Derivative of $y(x)=x^5$", size = 15)
show()

percent_error = abs(dydx3-actual)/actual
plot(x[:-2],percent_error[:-2]*100)
title("Percent Error", size = 15)
ylim(ymin=0, ymax=1)
show()
```

```python
def pointField(x,y,q,Xq,Yq):
    """Returns a tuple of the electric field caused by q at (Xq,Yq)"""
    k = 1.3806488e-23 #Nm^2/C^2
    Ex = (k*q*(x-Xq))/(((x-Xq)**2+(y-Yq)**2)**(0.5))
    Ey = (k*q*(y-Yq))/(((x-Xq)**2+(y-Yq)**2)**(0.5))
    return [Ex,Ey]
```
