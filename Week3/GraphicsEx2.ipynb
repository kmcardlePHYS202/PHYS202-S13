###Question 2

```python
x = arange(-pi,pi,0.01)
u = -cosh(x)**-2
i = 0+1j
l = [0, 0.5, 1.0, 2.0]
line = ['r-', 'r--', 'r-.', 'r:']
m = 0
for k in l:
    psik = ((i*k-tanh(x))/(i*k+1))*exp(i*k*x)
    plot(x, psik*psik.conjugate(), line[m], linewidth = 2.5)
    m+=1
plot(x, u, linewidth = 2.5)

xlabel("$x(rad)$", fontsize = 15)
ylabel("$U(x), \psi^2(x)$", fontsize = 15)
show()
```
