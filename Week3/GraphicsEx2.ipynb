###Question 2

```python
x = arange(-pi,pi,0.01)
u = -cosh(x)**-2
i = 0+1j
l = [0, 0.5, 1.0, 2.0]
line = ['r-', 'r--', 'r-.', 'r:']
m = 0
for k in l:
    psik = ((i*k-tanh(x))/(i*k+1))*exp(i*k*x)
    plot(x, psik*psik.conjugate(), line[m], linewidth = 2.5, label = "$\psi^2, k=%s$" %k)
    m+=1
plot(x, u, linewidth = 2.5, label = "$U(x)$")

xlabel("$x(rad)$", fontsize = 15)
ylabel("$U(x), \psi^2(x)$", fontsize = 15)

grid(True)
xticks(arange(-pi, pi+1, pi/2), ["$-{\pi}$", "$-{\pi}/2$", "$0$", "$+{\pi}/2$", "$+{\pi}$"])
legend(loc = 'lower left')
show()
```

###Question 3

```python
data = loadtxt("sunspots.txt", float)
x = data[:,0]
y = data[:,1]
v = data[:1001,0]
w = data[:1001,1]

plot(v,w, color = "red", linewidth = 0.7, linestyle="-", label = "$Number$ $of$ $Sunspots$")
def avg(array):
    total = 0
    count = 1
    for i in array:
        total += i
        count +=1
    return total/count

a = arange(1001)
for i in range(1001):
    a[i] = avg(w[:i])
    
plot(v, a, color = "blue", linewidth = 2.5, linestyle = ":", label = "$Average$")

xlabel("Time (years)", fontsize = 14)
ylabel("Number of Sunspots", fontsize = 14)
title("Sunspots Over Time", fontsize = 18)

grid(True)
xticks(arange(0,1000, 1000/(6)), [1749, 1762, 1776, 1790, 1804, 1818, 1832])
legend()
show()

```

###Question 4

```python
data = loadtxt("stars.txt",float)
x = data[:,0]
y = data[:,1]

scatter(x, y, s=5, color = "red", label = "$Magnitude$")
title("Hertzberg-Russel Diagram", size = 16)
ylabel("Magnitude", size = 14)
xlabel("Temperature (K)", size = 14)

xlim(xmax = 0, xmin = 16000)
ylim(ymax = -5, ymin = 20)

annotate('Main Sequence', xy = (6000, 5), xytext = (14000, 5), arrowprops=dict(facecolor='black', shrink=0.08),)
annotate('White Dwarfs', xy = (7200, 14), xytext = (15000, 15), arrowprops=dict(facecolor='black', shrink=0.08),)
annotate('Red Giants', xy = (5000, 1), xytext = (4000, -3), arrowprops=dict(facecolor='black', shrink=0.08),)

grid(True)
legend(loc = "lower center")
show()
```
