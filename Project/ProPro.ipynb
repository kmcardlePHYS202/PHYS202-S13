```python
#here is my actual code
from random import gauss

def outside(x,y):
    if y>sqrt((diameter/2)**2-x**2) or y<-sqrt((diameter/2)**2-x**2):
        return True
    return False

N = 10 #sample size
x = 1.3*r.random(N)-(diameter/2.) #random x value between -rad to +rad
y = 1.3*r.random(N)-(diameter/2.) #random y value between -rad to +rad
muon_E = zeros(N)
muon_v = zeros(N)

muon_mass = 105.7*(10**6) #eV/c^2
c = 3*(10**8) #m/s
v = lambda E: (c/(muon_mass))*sqrt(E**2 - (muon_mass)**2)

#make sure all points are within the circle
for i in range(N):
    muon_E[i] = gauss((4*(10**9)),(1.5*(10**9)))
    muon_v[i] = v(muon_E[i])
    while outside(x[i],y[i]):
        y[i] = 1.3*r.random(1)-(diameter/2.)
        x[i] = 1.3*r.random(1)-(diameter/2.)


        
figure(figsize=(5,5))
scatter(x,y)
show()

#to generate a sample of random incidence angles
theta = r.random(N) * 2*pi #along the plane of the atmosphere
phi = r.random(N) * pi     #angle from the normal to the plane of the atmosphere
```

```python
incident = [['x position:    ','y position:    ','theta:    ','\tphi:    ','\tmuon energy:    ','muon velocity:']]
for i in range(N):
    incident.append([x[i],y[i],theta[i],phi[i],muon_E[i],muon_v[i]])
for i in incident:
    print i[0],'    \t',i[1],'    \t',i[2],'    \t',i[3],'    \t',i[4],'    \t',i[5]

#energy in eV and velocity in m/s
```

```python
#declare 19 detectors [X, Y, Z, hit]
detectors = array([
[    0,      0, 0, 0],   #detector 1 at exactly center
[    0,   0.25, 0, 0],
[    0,    0.5, 0, 0],
[    0,  -0.25, 0, 0],
[    0,   -0.5, 0, 0],
[ 0.25,  0.125, 0, 0],
[ 0.25,  0.375, 0, 0],
[ 0.25, -0.125, 0, 0],
[ 0.25, -0.375, 0, 0],
[  0.5,      0, 0, 0],
[  0.5,   0.25, 0, 0],
[  0.5,  -0.25, 0, 0],
[-0.25,  0.125, 0, 0],
[-0.25,  0.375, 0, 0],
[-0.25, -0.125, 0, 0],
[-0.25, -0.375, 0, 0],
[ -0.5,      0, 0, 0],
[ -0.5,   0.25, 0, 0],
[ -0.5,  -0.25, 0, 0]])
```

```python
xdet=detectors[:,0]
ydet=detectors[:,1]

figure(figsize=(5,5))
sky_area = Circle((0,0),diameter/2,color='b',fill=False)
gca().add_artist(sky_area)
scatter(xdet,ydet)
title("Detector location")
xlim(-0.7,0.7)
ylim(-0.7,0.7)
show()
```

```python
#for test case where incedent is straight down on center
rad = height * tan(1*pi/180) #radius of hit area on ground
hit = zeros(detectors.shape[0])
for i in range(detectors.shape[0]):
    if (detectors[i,0]**2 + detectors[i,1]**2) <= rad**2:
        detectors[i,3] = 1
        hit[i] = 1

mask = (hit == 1)
area = Circle((0,0),rad,color = 'teal')

figure(figsize=(3,3))
sky_area = Circle((0,0),diameter/2,color='b',fill=False)
gca().add_artist(sky_area)
cir = gca().add_artist(area)
cir.set_zorder(0)
scatter(xdet,ydet,color='black')
scatter(xdet[mask],ydet[mask],color = 'red',lw=5)
title('Hit Detectors')
xlim(-0.7,0.7)
ylim(-0.7,0.7)
show()
```
