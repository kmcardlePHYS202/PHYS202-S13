Single Muon Detection Function
-

```python
#For a randomly placed particle straight down

def create_incident():
    c = 3*10**5 #km/s
    mc2 = 105.7*10**6 #eV
    vnew = lambda E: c*sqrt(1-(mc2/E)**2)#km/s

    rad = height * tan(1*pi/180) #radius of hit area on ground (1 degree radius)
    x0 = 1.3*r.random(1)-(diameter/2.) #random x value between -rad to +rad
    y0 = 1.3*r.random(1)-(diameter/2.) #random y value between -rad to +rad
    
    #theta = r.random(1) * 2*pi #along the plane of the atmosphere
    #phi = r.random(1) * pi     #angle from the normal to the plane of the atmosphere

    hit = zeros(detectors.shape[0])
    timing = zeros(detectors.shape[0])


    for i in range(detectors.shape[0]):
        if ((detectors[i,1]-x0)**2 + (detectors[i,2]-y0)**2) <= rad**2: #if its within the circle
            detectors[i,4] = 1
            hit[i] = 1
            
            muon_E = gauss((4*(10**9)),(1.5*(10**9))) #random energy based on gaussian for each muon
            muon_v = vnew(muon_E) #velocity based on random energy for each muon
        
            x = detectors[i,1]
            y = detectors[i,2]
            z = detectors[i,3]
            L = sqrt((x-x0)**2 + (y-y0)**2 + z**2)
            time = L/muon_v
        
            timing[i] = time

    mask = (hit == 1)
    hit_det = numdet[mask]
    #print hit_det
    #print timing[mask]
    timing = timing[mask]
    
    if len(timing)>0:
        min_t = min(timing)
        #print 'minimum time:',min_t,'s'

        for i in range(len(timing)):
            timing[i] = timing[i] - min_t + (0.05*r.random(1)*(timing[i]-min_t))
        #print timing
    
    if len(timing)>=4:
        hitormiss = 1
    else:
        hitormiss = 0
        #print 'MISS'
    result = [hitormiss]
    for i in range(len(timing)):
        result.append([int(hit_det[i]), timing[i]])
    result.append([float(x0),float(y0)]) #adds on the actual incident location to the result
    
    return result #returns if it was solid hit (3 or more) and which detectors were hit with what delta_t

print create_incident() 

```

Multilateration
-

I found the algorithms for the multilateration at this webiste: http://cs.fit.edu/~ryan/cse4051/projects/multilateration/mult.pdf

However it is only good for traingulating with 4 points

```python
def four_point_ml(det,times):

    c = 3*10**5 #km/s
    mc2 = 105.7*10**6 #eV
    vnew = lambda E: c*sqrt(1-(mc2/E)**2)#km/s

    Eavg = 4*(10**9) #eV   this is a rough estimate of what the energy is since we dont know the specific enerfy of each muon that hits
    Vavg = vnew(Eavg) #km/s

    #definitions... yay
    #since my code does not include hieght variance (z) there are a few lines that do not need to be included, they are commented out
    detA = detectors[det[0]-1]
    detB = detectors[det[1]-1]
    detC = detectors[det[2]-1]
    detD = detectors[det[3]-1]
    xi = detA[1]
    yi = detA[2]
    zi = detA[3]
    xj = detB[1]
    yj = detB[2]
    zj = detB[3]
    xk = detC[1]
    yk = detC[2]
    zk = detC[3]
    xl = detD[1]
    yl = detD[2]
    zl = detD[3]
    
    ti = times[0]
    tj = times[1]
    tk = times[2]
    tl = times[3]
    
    tij = ti-tj
    tik = ti-tk
    tkj = tk-tj
    tkl = tk-tl  #
    
    Rij = Vavg * tij
    Rik = Vavg * tik
    Rkj = Vavg * tkj
    Rkl = Vavg * tkl  #
    
    Xijy = Rij*(yk-yi) - Rik*(yj-yi)
    Xikx = Rik*(xj-xi) - Rij*(xk-xi)
    Xikz = Rik*(zj-zi) - Rij*(zk-zi)  ##
    Xkjy = Rkj*(yl-yk) - Rkl*(yj-yk)  #
    Xklx = Rkl*(xj-xk) - Rkj*(xl-xk)  #
    Xklz = Rkl*(zj-zk) - Rkj*(zl-zk)  #
    
    Si2 = xi**2 + yi**2 + zi**2
    Sj2 = xj**2 + yj**2 + zj**2
    Sk2 = xk**2 + yk**2 + zk**2
    Sl2 = xl**2 + yl**2 + zl**2  #
    
    Rij2xyz = Rij**2 + Si2 - Sj2
    Rik2xyz = Rik**2 + Si2 - Sk2
    Rkj2xyz = Rkj**2 + Sk2 - Sj2
    Rkl2xyz = Rkl**2 + Sk2 - Sl2  #
    
    A = Xikx/Xijy
    #B = Xikz/Xijy
    C = Xklx/Xkjy  #
    #D = Xklz/Xkjy
    
    E = (Rik*Rij2xyz - Rij*Rik2xyz)/(2*Xijy)
    F = (Rkl*Rkj2xyz - Rkj*Rkl2xyz)/(2*Xkjy)  #
    
    #G = (D-B)/(A-C)
    H = (F-E)/(A-C)
    
    #I = A*G + B
    J = A*H + E
    
    #K = Rik2xyz + 2*(xk-xi)*H + 2*(yk-yi)*J
    #L = 2*((xk-xi)*G + (yk-yi)*I + (zk-zi))
    
    #M = 4*Rik**2*(G**2 + I**2 + 1) - L**2
    #N = 8*Rik**2*(G*(xi-H) + I*(yi-J) + zi) + 2*L*K
    #O = 4*Rik**2*((xi-H)**2 + (yi-J)**2 + zi**2) - K**2
    
    def sign(x):
        if x<0:
            return -1
        if x==0:
            return 0
        if x>0:
            return 1
    
    #Q = N + sign(N)*sqrt(N**2 - 4*M*O)
    #z1 = Q/(2*M)
    #z2 = 2*O/Q
    #x0 = G*z+H
    #y0 = I*z+J
    x0 = H
    y0 = J
    
    #print Q
    return [x0,y0]
```
