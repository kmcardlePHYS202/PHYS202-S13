```python
dx = 0.25
x1 = arange(0,10,dx)
x = 1
y1 = zeros(x1.shape)
for i in range(x1.shape[0]):
    y1[i] = y1[i-1] + (x1[i]+2*y1[i-1])*dx
plot(x1,y1)
show()
print 'when x =',x1[4],' y =',y1[4] 
```

```python
dx = 0.02
x2 = arange(0,10,dx)
x = 1
y2 = zeros(x2.shape)
for i in range(x2.shape[0]):
    y2[i] = y2[i-1] + (x2[i]+2*y2[i-1])*dx
plot(x2,y2)
show()
print 'when x =',x2[50],' y =',y2[50] 
```

```python
x = arange(0,10,0.01)
f = lambda x: 0.25*exp(2*x) - 0.5*x - 0.25

plot(x1,y1, label='y in steps of 0.25')
plot(x2,y2, label='y in steps of 0.02')
plot(x,f(x), label='actual equation')
legend(loc = "upper left")
show()
```

it is easy to see that the 0.02 step size gives you a much more accurate result than the 0.25 step size

```python
#Euler-Cromer Method

#now that we have estimated values for them all we can use them in our calculations

dx = 0.25
x3 = arange(0,10,dx)
y3 = zeros(x3.shape)
for i in range(x3.shape[0]):
    y3[i] = y3[i-1] + (x3[i]+2*y2[i])*dx  #using the y[i] value from our previously calculated values
plot(x3,y3)
show()

dx = 0.02
x5 = arange(0,10,dx)
y5 = zeros(x5.shape)
for i in range(x5.shape[0]):
    y5[i] = y5[i-1] + (x5[i]+2*y2[i])*dx  #using the y[i] value from our previously calculated values
plot(x5,y5)
show()
```

```python
#Mid-Point Method

dx = 0.25
x4 = arange(0,10,dx)
y4 = zeros(x4.shape)
for i in range(x4.shape[0]):
    y4[i] = y4[i-1] + 0.5*((x4[i-1]+2*y2[i-1])+(x4[i]+2*y2[i]))*dx  #using the y[i] value from our previously calculated values
plot(x4,y4)
show()

dx = 0.02
x6 = arange(0,10,dx)
y6 = zeros(x6.shape)
for i in range(x6.shape[0]):
    y6[i] = y6[i-1] + 0.5*((x6[i-1]+2*y2[i-1])+(x6[i]+2*y2[i]))*dx  #using the y[i] value from our previously calculated values
plot(x6,y6)
show()
```

```python
plot(x3,f(x3)-y3, label='Euler-Cromer Method, step size 0.25')
plot(x5,f(x5)-y5, label='Euler-Cromer Method, step size 0.02')
plot(x4,f(x4)-y4, label='Mid-Point Method, step size 0.25')
plot(x6,f(x6)-y6, label='Mid-Point Method, step size 0.02')
legend(loc = 'upper left')
show()
```

you can see that of these the most accurate is the Mid-Point Method with a step size of 0.02
