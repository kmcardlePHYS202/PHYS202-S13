```python
a = -9.8
dt = 0.1
x = arange(0,10,dt)
v = zeros(x.shape)
v[0] = 0
for i in range(x.shape[0]):
    v[i] = v[i-1] + a*dt
plot(x,v)
show()
```

```python
r = zeros(x.shape)
r[0] = 0
for i in range(x.shape[0]):
    r[i] = r[i-1] + v[i]*dt
plot(x,r)
show()
```

```python
dx = 0.25
x1 = arange(0,10,dx)
x = 1
y1 = zeros(x1.shape)
for i in range(x1.shape[0]):
    y1[i] = y1[i-1] + (x1[i]+2*y1[i-1])*dx
plot(x1,y1)
show()
print 'when x =',x1[4],' y =',y1[4] 
```

```python
dx = 0.02
x2 = arange(0,10,dx)
x = 1
y2 = zeros(x2.shape)
for i in range(x2.shape[0]):
    y2[i] = y2[i-1] + (x2[i]+2*y2[i-1])*dx
plot(x2,y2)
show()
print 'when x =',x2[50],' y =',y2[50] 
```

```python
x = arange(0,10,0.01)
f = lambda x: 0.25*exp(2*x) - 0.5*x - 0.25

plot(x1,y1, label='y in steps of 0.25')
plot(x2,y2, label='y in steps of 0.02')
plot(x,f(x), label='actual equation')
legend(loc = "upper left")
show()
```
